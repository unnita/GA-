Logistic Recession with regularization to avoid overfitting 

```python
from sklearn.model_selection import train_test_split

X = df.loc[: , df.columns != 'diagnosis']
print(X.shape)

Y = df['diagnosis']
print(Y.shape)

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=4)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)

# checking uniform distribution of y labels
print(y_train.value_counts(), y_test.value_counts())
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)

from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression(C=10.0**-2,solver='liblinear')
logmodel.fit(X_train_std,y_train)

y_train_pred2 = logmodel.predict(X_train_std)
y_pred = logmodel.predict(X_test_std)

from sklearn.metrics import classification_report
print(classification_report(y_train,y_train_pred2))
```

```python
regularizationParam, auc, log_loss_difference = [], [], []

from sklearn.metrics import roc_auc_score

for c in np.arange(-5, 5):
    logmodel = LogisticRegression(C=10.0**c, random_state=0,solver='liblinear')
    logmodel.fit(X_train_std, y_train)
    regularizationParam = np.append(regularizationParam,c)
    auc = np.append(auc,roc_auc_score(y_test,logmodel.predict_proba(X_test_std)[:,1]))
```

```python
import matplotlib.pyplot as plt
plt.plot(regularizationParam, auc, color='blue', marker='x', label='mylabel')
```

```python
logmodel = LogisticRegression(C=10.0**-2, random_state=0,solver='liblinear')
logmodel.fit(X_train_std, y_train)
predictions = logmodel.predict(X_test_std)
```

```python
print(classification_report(y_test,predictions))
```

```python
coef_df = pd.DataFrame({'label': X_test.columns, 'abs coef': abs(logmodel.coef_[0]), 'coef': logmodel.coef_[0], 'odds ratio': np.exp(logmodel.coef_[0])}, columns=['label', 'abs coef', 'coef', 'odds ratio'])
coef_df.sort_values(by=['abs coef'], ascending=False)
```

Decision tree

I will use the cross validation to prevent overfitting

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

# Create the parameter grid 
param_grid = {
    'max_depth': range(2, 10),
    'max_features': range(1, 9),
    'min_samples_leaf': range(20, 100, 30),
    'min_samples_split': range(10, 50, 10),
    'criterion': ["entropy", "gini"]
}

n_folds = 5

# Initiate the grid search model
dtree = DecisionTreeClassifier()
grid_search = GridSearchCV(estimator = dtree, param_grid = param_grid, 
                          cv = n_folds, verbose = 1)
grid_search.fit(X_train,y_train)
grid_search.best_params_ , grid_search.best_score_
```

print("best accuracy", grid_search.best_score_)
print(grid_search.best_estimator_)

```python
tree = DecisionTreeClassifier(max_depth= 6,random_state=4,max_features= 5,min_samples_leaf=50,min_samples_split= 40)
tree = tree.fit(X_train,y_train)

y_train_pred3 = tree.predict(X_train)
y_pred3=tree.predict(X_test)

print(accuracy_score(y_train, y_train_pred3))
print(confusion_matrix(y_train, y_train_pred3))

print(accuracy_score(y_test, y_pred3))
print(confusion_matrix(y_test, y_pred3))
```

```python
tree_coef_df = pd.DataFrame({'label': X_test.columns, 'importances': tree.feature_importances_}, columns=['label', 'importances'])
tree_coef_df.sort_values(by=['importances'], ascending=False)
```

Comments on Decision Tree: Decision tree is a tree based used for classfication and I have used gridsearch used to avoid over fitting has performed moderately. looking at the coefficient we can see that the model has give features that are different from our previous analysis that contribute to model
